```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 50  # Sampling frequency (Hz)
f = 5  # Signal frequency (Hz)
bits = 3  # Number of quantization bits
#t = np.linspace(0, 1, fs, endpoint=False)  # Time vector (1 second), t = [0, 1/1000, 2/1000, ..., 998/1000, 999/1000]

# Time array for the continuous signal
t_cont = np.linspace(0, 1, 1000,endpoint=False) #t_cont = [0, 1/1000, 2/1000, ..., 1]

# Generate a sine wave (analog signal)
analog_signal = np.sin(2 * np.pi * f * t_cont)

# Sampling (take values at discrete points)
t_sampled = np.linspace(0, 1, fs, endpoint=False)
sampled_signal = np.sin(2 * np.pi * f * t_sampled)
#diladi gia kathe 20 points perno 1 point gia to sampled signal (1000/50 = 20)

# Quantization
M = 2**bits  # Number of quantization levels
quantized_signal = np.round(((sampled_signal + 1) / 2) * (M - 1)).astype(int)
#ta zyga strogilopoiountai pros ta kato, kai ta mona pros ta pano == banker's rounding

'''print("wavelength quantized_signal:")
for i in range(11):
    print (quantized_signal[i])'''
    
# Define binary levels (3-bit binary representation)
binary_levels = [format(i, '03b') for i in range(M)]
#range(levels) = [0,1,2,3,4,5,6,7]
#ex. i=5 format(5, '03b') = '101'  # to 03b simainei 3 bits se binary
#[format(i, '03b') for i in range(levels)] = ['000', '001', ... , '111']

print("wavelength sampled_signal:")
for i in range(11):
    print(sampled_signal[i])
print("\n\n")
print("wavelength binary_signal:")
for i in range(11):
    print (binary_levels[quantized_signal[i]])

quantized_signal = (quantized_signal / (M - 1)) * 1.75 - 0.875 #Scale back to the range [-0.875, 0.875]

print("\n\n")
print("wavelength quantized_signal:")
for i in range(11):
    print (quantized_signal[i])


# Plot results
plt.figure(figsize=(10, 10))

# 1. Plot Analog Signal
plt.subplot(3, 1, 1)
plt.plot(t_cont, analog_signal, color='blue', label="Analog Signal")
plt.title("Analog Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# 2. Plot Sampled Signal
plt.subplot(3, 1, 2)
plt.plot(t_cont, analog_signal, color='lightgray', linestyle='dashed', alpha=0.6, label="Analog Reference")
plt.scatter(t_sampled, sampled_signal, color='red', label="Sampled Signal")
plt.title("Sampled Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# 3. Plot Quantized Signal
plt.subplot(3, 1, 3)
plt.plot(t_cont, analog_signal, color='lightgray', linestyle='dashed', alpha=0.6, label="Analog Reference")
plt.scatter(t_sampled, quantized_signal, color='green', label="Quantized Signal")
plt.step(t_sampled, quantized_signal, where='mid', color='green', linestyle='--', label="PCM Output")
plt.title("Quantized Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
